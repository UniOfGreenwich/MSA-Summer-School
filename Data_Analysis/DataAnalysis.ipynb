### Showing Distributions by Plotting Histograms

A histogram represents the frequency distribution of variables in a dataset. Here we plot histograms to visualize data distributions.

```python
selected=['DHT11_Humidity']  # selecting the column of DHT11_Humidity 
plt.figure(figsize=(8, 8))
plt.hist(data[selected])     # drawing histogram
plt.xlabel('Humidity', fontsize=18)  # xlabel
plt.ylabel('Number of data points', fontsize=18) # ylabel
plt.show()
```

#### drawing a single figure with lots of subplots

```python
selected=['DHT11_Humidity','Motor_Temperature','RMS_Vibration','DHT11_Ambient_Temp']
counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Distributions of Some Propteries', fontsize=14, position=(0.5, 1.0))
for name_i in selected:
    sub = fig.add_subplot(3, 2, counter)   # we set 3 rows & 2 colums for showing subgraphs, the counter means the current i_th subgraph
    sub.set_title(name_i, fontsize=10)
    plt.hist(data[name_i]) 
    counter += 1
plt.subplots_adjust(wspace=1.0, hspace=1.0)
plt.tight_layout()
plt.show()
```

# Correlation among Variables

**Correlation** refers to a mutual relationship between two or more things or formally in statistics it refers to the interdependence of variables. For example, we can expect a positive correlation between temperature and ice cream sales when it’s hot, people buy more ice cream Or we can expect an inverse (or negative) correlation between temperature and sales of gloves when it’s hot, people buy fewer gloves.

The usual way to visualise correlation is with a scatter plot take any two variables, e.g. A & B, with common index values (e.g. dates)
for each index value (e.g. each date), plot A’s value along the x-axis against B’s value on the y-axis with a small dot at the location (A, B).

If you get a diagonal line sloping upwards, left to right, the variables are **positively** correlated a diagonal line sloping downwards, left to right, the variables are **negatively** (inversely) correlated randomly scattered dots, the variables are not correlated.


```python
# scatter plot 
selected = ['DHT11_Humidity', 'Motor_Temperature']
plt.figure(figsize=(8, 8))
plt.scatter(data['DHT11_Humidity'], data['Motor_Temperature']) # draw a scatter plot 
plt.title('DHT11_Humidity' + ' vs ' + 'Motor_Temperature', fontsize=20)
plt.xlabel(name_i, fontsize=18)
plt.ylabel(name_j, fontsize=18)
plt.show()
```

### HeatMap correlation


Even with Pandas calculating correlations we still need to visualise them somehow for 25 variables, Pandas calculates 25 x 24 = 600 correlation coefficients. The standard way to do this is with a heat map, a coloured chart where blue represents cold (-1) and red represents hot (+1).It’s certainly possible to do this in MatPlotLib but it’s much easier using **Seaborn**, a high level visualisation library built on top of Pandas and MatPlotLib.

```python
# %pip install seaborn
import seaborn as sns
selected = ['DHT11_Humidity', 'Motor_Temperature', 'RMS_Vibration','DHT11_Ambient_Temp']
plt.figure(figsize=(10, 10))
corr = data[selected].corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(220, 20, n=200),
    square=True, annot=True, annot_kws={"size": 8})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45,
    horizontalalignment='right')
plt.show()

```

### Trendlines

A trend is a general direction in which something is developing or changing. We can calculate a **trendline** for each time-series a line showing whether the general trend of the time-series is increasing over time or decreasing or flat.  The formal name for this is **linear regression**  because we compute a straight line through the data points.

Pandas doesn’t have a method to calculate trendlines instead we use **NumPy** to do this.

```python
#%pip install numpy
import numpy as np
plt.figure(figsize=(8, 8))
plt.plot(data['RMS_Vibration'], linewidth=0.5)
plt.gca().set_prop_cycle(None)
x = np.arange(len(data['RMS_Vibration']))
z = np.polyfit(x, data['RMS_Vibration'], 1)  # fit the data
trend = np.poly1d(z)
plt.plot(data.index, trend(x), linestyle='--',color='red')
plt.xlabel('Time', fontsize=18)
plt.ylabel('RMS_Vibration', fontsize=18)
plt.title('RMS_Vibration'
    + ' with Trendlines', fontsize=20)
plt.legend('RMS_Vibration', loc=2)
plt.show()
```

NumPy fits a polynomial curve to the time-series using **polyfit()**.  the '1' indicates that it is linear (a straight line is a polynomial of degree 1. We “save” the trendline NumPy has calculated in a function called trend Plot the trendline against the dates (i.e. the index of the dataframe) .
